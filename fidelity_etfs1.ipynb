```python
import pandas as pd
import requests
import pprint
import yfinance as yf

from tqdm import tqdm, tqdm_notebook
tqdm_notebook().pandas()

import bs4 as bs
import urllib.request
import ssl
ssl._create_default_https_context = ssl._create_unverified_context
```

```python
def extract_table(url):
    
    source = urllib.request.urlopen(url).read()
    soup = bs.BeautifulSoup(source,'lxml')
    
    data = []
    table = soup.find('table', attrs={'class':'table mm-mobile-table table-module2 table-default table-striped table-hover table-pagination'})
    table_body = table.find('tbody')

    rows = table_body.find_all('tr')
    for row in rows:
        cols = row.find_all('td')
        cols = [ele.text.strip() for ele in cols]
        data.append([ele for ele in cols if ele]) # Get rid of empty values
        
    tmp = pd.DataFrame(data)
    page_etfs = tmp.iloc[:,:8]
    page_etfs.columns = ["Symbol", "ETF Name", "Asset Class", "Total Assets ($MM)", "YTD Price Change",
                      "Avg. Daily Volume", "Previous Closing Price", "1-Day Change"]

    return page_etfs
```

```python
urls = ["https://etfdb.com/etfs/issuers/fidelity/#etfs&sort_name=assets_under_management&sort_order=desc&page=1",
        "https://etfdb.com/etfs/issuers/fidelity/#etfs&sort_name=assets_under_management&sort_order=desc&page=2"]

df = pd.concat([extract_table(url) for url in urls])
```

```python
df = df.reset_index()
del df["index"]
df.head()
```

```python
def get_details(symbol):
    
    tmp = yf.Ticker(symbol)
    return tmp.info

df["details"] = df["Symbol"].progress_apply(get_details)
```

```python
df.to_pickle("serialized/fidelity_etf.pql")
```
