```python
import pandas as pd
from tqdm import tqdm, tqdm_notebook
tqdm_notebook().pandas()
```

```python
df = pd.read_pickle("serialized/fidelity_etf.pql")
df.head()
```

```python
tmp = df["details"][0]
keys_to_remove = ["exchange", "exchangeTimezoneName", "exchangeTimezoneShortName", "isEsgPopulated", "gmtOffSetMilliseconds", "messageBoardId", "market", "bondHoldings",
                  "bondRatings", "logo_url"]

new_columns = [x for x in tmp.keys() if x not in keys_to_remove]
for col in new_columns:
    df[col] = df["details"].apply(lambda x: x[col] if col in x else "")
```
